```python
A = {1, 2, 3, 4, 5, 6, "один", "два"}
B = {4, 5, 6, 7, 8, 9, "два", "три"}
C = {7, 8, 9, 10, 11, 12, "три", "четыре"}
```

```python
# Объединение
print(A|B)
print(A|C)
print(B|C)
```

```python
# Пересечение
print(A&B)
print(A&C)
print(B&C)
```

```python
# Разность
print(A-B)
print(A-C)
print(B-C)
print(B-A)
print(C-B)
print(C-A)
```

```python
#Симметричную разность
print(A^B)
print(A^C)
print(B^C)
```
